```python
# ======================================================
# 🧪 CBTis 128
# 📚 Programación Orientada a Objetos 2026
# 🤖 Especialidad: Inteligencia Artificial
# 👨‍🏫 Docente: Eliseo Nava
# 👨‍🎓 Alumno: Marco Morquecho
# 🧪 Proyecto: Diccionario de Científicos
# ======================================================

# =========================
# 🧪 Diccionario vacío
# =========================
diccionario_cientificos = {}

# =========================
# ➕ Agregar científico
# =========================
def agregar_cientifico():

    id_cientifico = input("Ingresa ID del científico: ")

    rama = input("Ingresa rama de ciencia: ")

    experimento = input("Ingresa experimento activo: ")

    patente = input("Ingresa patente: ")

    laboratorio = input("Ingresa equipo de laboratorio: ")

    presupuesto = input("Ingresa presupuesto de investigación: ")

    tesis = input("Ingresa tesis dirigidas: ")

    diccionario_cientificos[id_cientifico] = {

        "Rama": rama,
        "Experimento": experimento,
        "Patente": patente,
        "Laboratorio": laboratorio,
        "Presupuesto": presupuesto,
        "Tesis": tesis
    }

    print("Científico agregado correctamente")

    listar_cientificos()

# =========================
# 📜 Listar científicos
# =========================
def listar_cientificos():

    print("\n===================================================================================================")
    print("ID | RAMA | EXPERIMENTO | PATENTE | LABORATORIO | PRESUPUESTO | TESIS")
    print("===================================================================================================")

    for key, value in diccionario_cientificos.items():

        print(key, "|",
              value["Rama"], "|",
              value["Experimento"], "|",
              value["Patente"], "|",
              value["Laboratorio"], "|",
              value["Presupuesto"], "|",
              value["Tesis"])

    print("===================================================================================================")

# =========================
# 🔍 Buscar científico
# =========================
def buscar_cientifico():

    buscar = input("Ingresa ID del científico: ")

    if buscar in diccionario_cientificos:

        print("\n===================================================================================================")
        print("ID | RAMA | EXPERIMENTO | PATENTE | LABORATORIO | PRESUPUESTO | TESIS")
        print("===================================================================================================")

        value = diccionario_cientificos[buscar]

        print(buscar, "|",
              value["Rama"], "|",
              value["Experimento"], "|",
              value["Patente"], "|",
              value["Laboratorio"], "|",
              value["Presupuesto"], "|",
              value["Tesis"])

        print("===================================================================================================")

    else:
        print("Científico no encontrado")

# =========================
# ❌ Borrar científico
# =========================
def borrar_cientifico():

    borrar = input("Ingresa ID del científico a borrar: ")

    if borrar in diccionario_cientificos:

        del diccionario_cientificos[borrar]

        print("Científico eliminado correctamente")

        listar_cientificos()

    else:
        print("Científico no encontrado")

# =========================
# ✏️ Actualizar científico
# =========================
def actualizar_cientifico():

    actualizar = input("Ingresa ID del científico: ")

    if actualizar in diccionario_cientificos:

        rama = input("Nueva rama de ciencia: ")

        experimento = input("Nuevo experimento activo: ")

        patente = input("Nueva patente: ")

        laboratorio = input("Nuevo equipo de laboratorio: ")

        presupuesto = input("Nuevo presupuesto: ")

        tesis = input("Nuevas tesis dirigidas: ")

        diccionario_cientificos[actualizar] = {

            "Rama": rama,
            "Experimento": experimento,
            "Patente": patente,
            "Laboratorio": laboratorio,
            "Presupuesto": presupuesto,
            "Tesis": tesis
        }

        print("Científico actualizado correctamente")

        listar_cientificos()

    else:
        print("Científico no encontrado")

# =========================
# 📌 Menú principal
# =========================
while True:

    print("\n========= MENÚ CIENTÍFICOS 🧪 =========")
    print("1. Agregar científico")
    print("2. Listar científicos")
    print("3. Buscar científico")
    print("4. Borrar científico")
    print("5. Actualizar científico")
    print("6. Salir")

    opcion = int(input("Selecciona una opción: "))

    if opcion == 1:
        agregar_cientifico()

    elif opcion == 2:
        listar_cientificos()

    elif opcion == 3:
        buscar_cientifico()

    elif opcion == 4:
        borrar_cientifico()

    elif opcion == 5:
        actualizar_cientifico()

    elif opcion == 6:
        print("Programa finalizado 🧪")
        break

    else:
        print("Opción no válida")
```
